# **2030 SSP1-2.6 scenario**

```python
Cmip6_126 = xr.open_dataset(r'//content/drive/My Drive/Ester AIMS/AIMS project/SSP1-2.6_dataset.nc')
# Cmip6_126
```

```python
#computing bumber of dys when maximum temperatures is above 30 degrees celsius
Cmip6_126 = xclim.indices.tx_days_above(Cmip6_126.tasmax, thresh="30 degC", freq="MS")
```

```python
yearly_mean = Cmip6_126.groupby('time.year').mean(keep_attrs=True)
```

```python
Cmip6_126_2030 = yearly_mean.sel(year=2030, method='nearest')

fig = plt.figure(figsize=(11,8.5))

# Set the axes using the specified map projection
ax=plt.axes(projection=ccrs.PlateCarree())
ax.add_feature(cf.STATES.with_scale('10m'), zorder=3, linewidth=0.5, edgecolor='black')
#ax.set_extent([-20, 60, -40, 45], crs=ccrs.PlateCarree())
ax.set_extent([33, 42, -5, 5], crs=ccrs.PlateCarree())
# Make a filled contour plot
cs = ax.contourf(Cmip6_126_2030['lon'], Cmip6_126_2030['lat'], Cmip6_126_2030,
            transform = ccrs.PlateCarree(),cmap='RdBu_r',extend='both')


# Add coastlines
ax.coastlines()
ax.add_feature(cf.BORDERS)
ax.add_feature(cf.RIVERS)
ax.add_feature(cf.LAKES)
ax.add_feature(cf.OCEAN, zorder=1, edgecolor='white')

#title of the map
ax.set_title('2030 Maximum surface temperature for SSP1-2.6 scenario', fontsize=12, fontweight='bold')
# Add colorbar
cbar = plt.colorbar(cs,shrink=1.0,orientation='vertical',label='Number of days temperature > 30 degrees')
```
